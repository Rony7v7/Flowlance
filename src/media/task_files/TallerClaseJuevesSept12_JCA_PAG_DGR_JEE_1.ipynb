# 6. Volumen de un anillo para un adulto promedio

**1. Obtención de un valor estándar de volumen**

Para determinar un valor estándar del volumen de un anillo para un adulto promedio, consideramos la masa típica de un anillo y la densidad del material del que está hecho.

Supongamos que un anillo típico de oro pesa aproximadamente $4 \, \text{g}$. La densidad del oro es aproximadamente $19.32 \, \text{g/cm}^3$.

Convertimos la densidad a $\text{g/mm}^3$:

$$
19.32 \, \frac{\text{g}}{\text{cm}^3} = 19.32 \, \frac{\text{g}}{\text{cm}^3} \times \frac{1 \, \text{cm}^3}{1000 \, \text{mm}^3} = 0.01932 \, \frac{\text{g}}{\text{mm}^3}
$$

Ahora, calculamos el volumen del anillo:

$$
V = \frac{\text{Masa}}{\text{Densidad}} = \frac{4 \, \text{g}}{0.01932 \, \frac{\text{g}}{\text{mm}^3}} \approx 207 \, \text{mm}^3
$$



**2. Cálculo de $a$ y $b$ utilizando el volumen estándar**

Utilizamos la fórmula del volumen del anillo:

$$
V = \frac{4\pi}{3} \left( b^2 - a^2 \right)^{3/2}
$$

Despejamos $\left( b^2 - a^2 \right)^{3/2}$:

$$
\left( b^2 - a^2 \right)^{3/2} = \frac{3V}{4\pi}
$$

Sustituyendo $V = 207 \, \text{mm}^3$:

$$
\left( b^2 - a^2 \right)^{3/2} = \frac{3 \times 207}{4\pi} = \frac{621}{4\pi} \approx 49.43 \, \text{mm}^3
$$

Despejamos $b^2 - a^2$:

$$
b^2 - a^2 = \left( 49.43 \, \text{mm}^3 \right)^{2/3} \approx 13.25 \, \text{mm}^2
$$

Tomamos un valor promedio para el radio interior del anillo, $a = 9 \, \text{mm}$. Entonces:

$$
\begin{align*}
b^2 &= a^2 + 13.25 \\
&= (9 \, \text{mm})^2 + 13.25 \, \text{mm}^2 \\
&= 81 \, \text{mm}^2 + 13.25 \, \text{mm}^2 \\
&= 94.25 \, \text{mm}^2
\end{align*}
$$

Calculamos $b$:

$$
b = \sqrt{94.25 \, \text{mm}^2} \approx 9.71 \, \text{mm}
$$

El grosor del anillo es:

$$
\text{Grosor} = b - a = 9.71 \, \text{mm} - 9 \, \text{mm} = 0.71 \, \text{mm}
$$

Este grosor es menor que el típico para un anillo. Por lo tanto, ajustamos el volumen estándar a un valor mayor, por ejemplo, $V = 500 \, \text{mm}^3$.

Recalculamos:

$$
\left( b^2 - a^2 \right)^{3/2} = \frac{3V}{4\pi} = \frac{3 \times 500}{4\pi} = \frac{1500}{4\pi} \approx 119.37 \, \text{mm}^3
$$

Despejamos $b^2 - a^2$:

$$
b^2 - a^2 = \left( 119.37 \, \text{mm}^3 \right)^{2/3} \approx 27.87 \, \text{mm}^2
$$

Con $a = 9 \, \text{mm}$:

$$
\begin{align*}
b^2 &= a^2 + 27.87 \\
&= 81 \, \text{mm}^2 + 27.87 \, \text{mm}^2 \\
&= 108.87 \, \text{mm}^2
\end{align*}
$$

Calculamos $b$:

$$
b = \sqrt{108.87 \, \text{mm}^2} \approx 10.44 \, \text{mm}
$$

El grosor del anillo ahora es:

$$
\text{Grosor} = b - a = 10.44 \, \text{mm} - 9 \, \text{mm} = 1.44 \, \text{mm}
$$

Este grosor es más adecuado para un anillo promedio.



**Conclusión:** Para obtener el volumen estándar de un anillo de adulto promedio, podemos utilizar $a = 9 \, \text{mm}$ y $b = 10.44 \, \text{mm}$, lo que resulta en un grosor razonable y un volumen consistente con un anillo típico.

Entonces:

La integral triple a resolver es:

$$
\int_0^\pi \int_0^{2\sin(\theta)} \int_0^{\sqrt{4 - r^2}} r \, dz \, dr \, d\theta
$$

\section*{Paso 1: Resolver la Integral Más Interna}

La integral más interna con respecto a $z$ es:

$$
\int_0^{\sqrt{4 - r^2}} r \, dz
$$

Dado que $r$ es constante respecto a $z$, podemos sacarlo fuera de la integral:

$$
r \int_0^{\sqrt{4 - r^2}} dz = r \left[ z \right]_0^{\sqrt{4 - r^2}} = r \sqrt{4 - r^2}
$$

\section*{Paso 2: Resolver la Integral Intermedia}

Ahora resolvemos la integral con respecto a $r$:

$$
\int_0^{)} r \sqrt{4 - r^2} \, dr
$$

Usamos la sustitución $u = 4 - r^2$, lo que implica que $du = -2r \, dr$. Entonces la integral se convierte en:

$$
-\frac{1}{2} \int_4^{4 - 4\sin^2(\theta)} \sqrt{u} \, du
$$

La integral de $\sqrt{u}$ es:

$$
\int \sqrt{u} \, du = \frac{2}{3} u^{3/2}
$$

Evaluamos en los límites:

$$
-\frac{1}{2} \cdot \frac{2}{3} \left[ u^{3/2} \right]_4^{4 - 4\sin^2(\theta)} = -\frac{1}{3} \left[ \left(4 - 4\sin^2(\theta)\right)^{3/2} - 4^{3/2} \right]
$$

Simplificamos:

$$
4^{3/2} = 8
$$
$$
(4 - 4\sin^2(\theta))^{3/2} = 8\cos^3(\theta)
$$

Por lo tanto:

$$
-\frac{1}{3} \left[ 8\cos^3(\theta) - 8 \right] = \frac{8}{3} \left( 1 - \cos^3(\theta) \right)
$$

\section*{Paso 3: Resolver la Integral Externa}

Ahora integramos con respecto a $\theta$:

$$
\int_0^\pi \frac{8}{3} \left( 1 - \cos^3(\theta) \right) \, d\theta
$$

Dividimos la integral en dos partes:

1. La primera parte es:

$$
\frac{8}{3} \int_0^\pi 1 \, d\theta = \frac{8}{3} \cdot \pi
$$

2. La segunda parte es la integral de $\cos^3(\theta)$. Sabemos que la integral de $\cos^3(\theta)$ sobre el intervalo $[0, \pi]$ es cero:

$$
\int_0^\pi \cos^3(\theta) \, d\theta = 0
$$

#Resultado Final

Sumando las dos partes, obtenemos:

$$
\frac{8}{3} \cdot \pi + 0 = \frac{8\pi}{3}
$$

Por lo tanto, el valor de la integral triple es:

$$
\boxed{\frac{8\pi}{3}}
$$
